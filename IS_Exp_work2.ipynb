```python
import pandas as pd

pd.set_option('display.max_rows', None)  # 显示所有行
data = pd.read_csv("D:\PycharmProjects\informationSecurity\data\salaries_cyber.csv")
data.head()
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

fig, axes = plt.subplots(1, 1, num="stars", figsize=(8, 6))

plt.style.use('seaborn-darkgrid')

palette = plt.get_cmap('tab20c')
plt.subplot(1, 1, 1)

ax1 = sns.histplot(data['salary_in_usd'], kde=True, bins=579, shrink=1, color=palette.colors[0],
                   edgecolor=palette.colors[-1])

plt.show()
```

```python
ax = sns.boxplot(x='experience_level', y='salary_in_usd', data=data, notch=True)

ax = sns.swarmplot(x='experience_level', y='salary_in_usd', data=data, palette="Set2",linewidth=0.5, size=2)
```

```python
ax = sns.boxplot(x='company_size', y='salary_in_usd', data=data, notch=True)

ax = sns.swarmplot(x='company_size', y='salary_in_usd', data=data, palette="Set2",linewidth=0.5, size=2)
```

```python
# 计算各列值的频率
experience_freq = data['experience_level'].value_counts()
company_size_freq = data['company_size'].value_counts()
employment_type_freq = data['employment_type'].value_counts()

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制第一个直方图
plt.subplot(1, 3, 1)
plt.bar(experience_freq.index, experience_freq.values, color='purple')
plt.title('Experience Level')
plt.xlabel('Level')
plt.ylabel('Frequency')

# 绘制第二个直方图
plt.subplot(1, 3, 2)
plt.bar(company_size_freq.index, company_size_freq.values, color='red')
plt.title('Company Size')
plt.xlabel('Size')
plt.ylabel('Frequency')

# 绘制第三个直方图
plt.subplot(1, 3, 3)
plt.bar(employment_type_freq.index, employment_type_freq.values, color='green')
plt.title('Employment Type')
plt.xlabel('Type')
plt.ylabel('Frequency')

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```

```python
# 绘制热力图
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')

# 显示图形
plt.show()
```
